```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt


np.random.seed(42)
data_normal = np.random.normal(0, 1, (100, 2))  # داده‌های معمولی
data_outliers = np.random.uniform(low=-4, high=4, size=(10, 2))  # داده‌های پرت
data = np.vstack((data_normal, data_outliers))  # ترکیب داده‌ها

# DataFrame Converting
df = pd.DataFrame(data, columns=['Feature1', 'Feature2'])

# Isolation Forest
model = IsolationForest(contamination=0.1, random_state=42)
df['Outlier'] = model.fit_predict(df[['Feature1', 'Feature2']])


outliers = df[df['Outlier'] == -1]
clean_data = df[df['Outlier'] == 1]


print("تعداد داده‌ها:")
print(f"کل داده‌ها: {len(df)}")
print(f"داده‌های پرت: {len(outliers)}")
print(f"داده‌های تمیز: {len(clean_data)}")


plt.scatter(clean_data['Feature1'], clean_data['Feature2'], label='Clean Data', c='blue')
plt.scatter(outliers['Feature1'], outliers['Feature2'], label='Outliers', c='red')
plt.legend()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Isolation Forest - Outlier Detection')
plt.show()
```
